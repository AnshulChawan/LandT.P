```python
import os
import shutil
from ultralytics import YOLO
```

```python
# Load your trained YOLO model
model = YOLO("/content/drive/MyDrive/best.pt")  # Replace with the actual path to your model
```

```python
# Path to the test dataset with mixed images and labels
mixed_test_path = "/content/drive/MyDrive/test"
```

```python
# Create separate directories for images and labels
test_path_img = os.path.join(mixed_test_path, "images")
test_path_label = os.path.join(mixed_test_path, "labels")
os.makedirs(test_path_img, exist_ok=True)
os.makedirs(test_path_label, exist_ok=True)
```

```python
# Separate images and labels into their respective directories
for filename in os.listdir(mixed_test_path):
    if filename.endswith(".jpg"):
        shutil.move(os.path.join(mixed_test_path, filename), test_path_img)
    elif filename.endswith(".txt"):
        shutil.move(os.path.join(mixed_test_path, filename), test_path_label)
```

```python
results=model('/content/drive/MyDrive/test/images')
```

```python
from PIL import Image
import numpy as np
```

```python
i=1
```

```python
for r in results:
    im_array = r.plot()  # plot a BGR numpy array of predictions
    im = Image.fromarray(im_array[..., ::-1])  # RGB PIL image
    im.show()  # show image

    # Generate a unique filename with a counter
    filename = f"tested_{i}.jpg"
    i+=1

    # Save the image to the specified path
    im.save(os.path.join('/content/drive/MyDrive/BIGPRO/OUTPUT/Cars/tested', filename))
```
